```python
import os
import joblib
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error

# 1. Otteniamo la cartella corrente (perfetto per Jupyter Notebook)
current_dir = os.getcwd()

# 2. Sistemiamo il percorso per puntare sempre dentro "xgboost"
if os.path.basename(current_dir) == "xgboost":
    # Se il notebook è già dentro xgboost, restiamo qui
    base_dir = current_dir
else:
    # Se il notebook è nella cartella principale, entriamo in xgboost
    base_dir = os.path.join(current_dir, "xgboost")

# 3. Definiamo la cartella resource
resource_dir = os.path.join(base_dir, "resources")

# Percorsi completi per input e output
model_path = os.path.join(resource_dir, "social_matcher_model.pkl")
X_test_path = os.path.join(resource_dir, "X_test.pkl")
y_test_path = os.path.join(resource_dir, "y_test.pkl")

required_files = [model_path, X_test_path, y_test_path]

# Verifica
for file in required_files:
    if not os.path.exists(file):
        raise FileNotFoundError(f"File mancante!\nPercorso cercato: '{file}'\nImpossibile procedere.")

# Caricamento e previsioni
model = joblib.load(model_path)
X_test = joblib.load(X_test_path)
y_test = joblib.load(y_test_path)
predictions = np.clip(model.predict(X_test), 0, 100)

# Metriche
mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))

print("\n==================================")
print("      PRESTAZIONI DEL MODELLO     ")
print("==================================")
print(f"Errore Medio Assoluto (MAE) : {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print("==================================")
```
